```python
turnout_df = pd.read_csv('2020_turnout_reformatted.csv')
turnout_df.head()
```

```python
plt.figure(figsize=(4,15))

plt.barh(range(len(turnout_df)), 
        turnout_df['VEP Turnout Rate (Total Ballots Counted)'].sort_values())

plt.yticks([ x for x in range(len(turnout_df)) ], 
           turnout_df['State'])
plt.xlim(min(turnout_df['VEP Turnout Rate (Total Ballots Counted)']) - .1)

plt.tight_layout(True)
```

```python
turnout_df['% in Prison'] = turnout_df['Prison']/turnout_df['Voting-Eligible Population (VEP)']


plt.scatter(turnout_df['% in Prison'],
            turnout_df['VEP Turnout Rate (Total Ballots Counted)'])
plt.xlim(min(turnout_df['% in Prison'])-.001,max(turnout_df['% in Prison'])+.001)
plt.xlabel("% on Population in Prison")
plt.ylabel("Voter Turnout")
plt.show()
```

```python
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(turnout_df['% in Prison'],
                turnout_df['VEP Turnout Rate (Total Ballots Counted)'])
```

```python
plt.scatter(turnout_df['% in Prison'],
            turnout_df['VEP Turnout Rate (Total Ballots Counted)'])
plt.xlim(min(turnout_df['% in Prison'])-.001,max(turnout_df['% in Prison'])+.001)
plt.xlabel("% on Population in Prison")
plt.ylabel("Voter Turnout")

plt.plot(turnout_df['% in Prison'],
         [ (slope * x + intercept) for x in turnout_df['% in Prison'] ],
        'r')

print(f"The r-squared (Coefficient of Detirmination) is {rvalue**2}, ")
plt.show()
```
